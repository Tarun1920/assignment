Matplotlib is a popular and widely-used Python library for creating static, animated, and interactive visualizations in a 2D format. It provides a high-quality interface for drawing a variety of plots and charts, making it a fundamental tool for data visualization and scientific computing in Python.

Line Plot: Used to display data points connected by straight lines, often used to show trends over time or continuous data.

Scatter Plot: A plot of individual data points, often used to visualize the relationship between two variables.

Bar Chart: Displays data using bars of different heights, useful for comparing categories or showing the distribution of discrete data.

Histogram: Represents the distribution of a single variable by dividing it into bins and counting the number of data points in each bin.

Pie Chart: A circular chart that represents data as slices of a pie, typically used to show the composition of a whole in terms of its parts

A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional plane. It is typically used to show the relationship between two variables, making it useful for identifying patterns, trends, and potential outliers in data.



```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(3)

# Generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add a title, xlabel, and ylabel
plt.title("Scatter Plot of x and y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
```

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots in rows and columns, making it useful for visualizing multiple datasets or aspects of data in a single figure. This function takes three arguments: the number of rows, the number of columns, and the current subplot index.



```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2x2 grid of subplots
plt.figure(figsize=(10, 8))

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)  # First subplot
plt.plot(x1, y1, label='Line 1')
plt.title('Line 1')

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)  # Second subplot
plt.plot(x2, y2, label='Line 2')
plt.title('Line 2')

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)  # Third subplot
plt.plot(x3, y3, label='Line 3')
plt.title('Line 3')

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)  # Fourth subplot
plt.plot(x4, y4, label='Line 4')
plt.title('Line 4')

# Add a legend
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the entire figure
plt.show()
```

A bar plot (or bar chart) is a type of data visualization that uses rectangular bars to represent data values. It is used to compare and display the values of different categories or groups, making it suitable for visualizing categorical or discrete data. Bar plots are often used to show the relationship between different items and their corresponding values, allowing for easy comparison and identification of trends.

Bar plots can be either vertical or horizontal. In a vertical bar plot, the bars extend upwards from the x-axis, while in a horizontal bar plot, the bars extend from the y-axis. The choice between a vertical and a horizontal bar plot depends on the nature of the data and the visual preferences.



```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel("Company")
plt.ylabel("Profit (in billions USD)")
plt.title("Company Profits (Vertical Bar Plot)")
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightcoral')
plt.xlabel("Profit (in billions USD)")
plt.ylabel("Company")
plt.title("Company Profits (Horizontal Bar Plot)")
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It is used to visualize the central tendency, spread, and shape of the data's distribution, as well as to identify potential outliers. Box plots are particularly useful for comparing the distribution of multiple datasets and identifying patterns or differences between them.



```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=labels, vert=False, patch_artist=True)

plt.xlabel("Value")
plt.ylabel("Box")
plt.title("Box Plot of Box 1 and Box 2")
plt.grid(True, axis='x', linestyle='--', alpha=0.6)

plt.show()
```
