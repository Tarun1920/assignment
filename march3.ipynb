Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five common types of plots you can create with Seaborn and their typical use cases:

Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two numerical variables. They help in identifying patterns, correlations, and outliers in data.
Bar Plot:

Use: Bar plots are used to display categorical data with rectangular bars. They are useful for comparing categories or showing the distribution of a categorical variable.
Histogram:

Use: Histograms are used to visualize the distribution of a single numerical variable. They display the frequency or probability of different values within a data set.
Box Plot (Box-and-Whisker Plot):

Use: Box plots are useful for visualizing the distribution and spread of data. They show the median, quartiles, and potential outliers in a dataset, making them great for detecting skewness and identifying central tendencies.
Heatmap:

Use: Heatmaps are effective for visualizing complex data and correlations between variables. They use color to represent values in a matrix, making it easy to spot patterns and relationships in large datasets, especially in fields like data analysis and genomics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset from Seaborn
fmri = sns.load_dataset("fmri")

# Create a line plot using Seaborn
sns.set(style="whitegrid")  # Optional: Set the style of the plot

# Plot the line plot with "timepoint" on the x-axis and "signal" on the y-axis, differentiating by "event" and "region"
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Add labels and a legend
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal Over Time")
plt.legend(title="Event")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset from Seaborn
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot the box plot for 'age' categorized by 'pclass'
sns.boxplot(data=titanic, x="pclass", y="age", ax=axes[0])
axes[0].set_title("Age by Pclass")

# Plot the box plot for 'fare' categorized by 'pclass'
sns.boxplot(data=titanic, x="pclass", y="fare", ax=axes[1])
axes[1].set_title("Fare by Pclass")

# Show the plots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset from Seaborn
diamonds = sns.load_dataset("diamonds")

# Create a histogram with hue for the 'cut' column
sns.histplot(data=diamonds, x="price", hue="cut", element="step", common_norm=False)

# Add labels and a legend
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")
plt.legend(title="Cut")

# Show the plot
plt.show()
```
